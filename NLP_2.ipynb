# Minimum Edit Distance

```python
import numpy as np
```

```python
text1='deal'
text2='leda'
```

```python
dist=np.zeros((len(text1)+1,len(text2)+1))
dist[0,:]=np.arange(len(text2)+1)
dist[:,0]=np.arange(len(text1)+1)
print(dist)
```

# Dynamic programming to calculate minimum edit distance

```python
for i in range(1,len(text1)+1):
    for j in range(1,len(text2)+1):
        if text1[i-1]==text2[j-1]:
            cost=0
        else:
          dist[i,j]=min(dist[i-1,j]+1,
                        dist[i,j-1]+1,
                        dist[i-1,j-1]+2)
print(dist)
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk.tree import Tree

# Example CFG বানানো
non_recursive_cfg=nltk.CFG.fromstring("""
   S -> A'a'
   A -> 'b' | 'c'
   """)

# Input text
text = "b a"
words = text.split()
print(words)   # ['b', 'a']

# Parser তৈরি
parser = nltk.ChartParser(non_recursive_cfg)

# Parse করে tree পাওয়া
for tree in parser.parse(words):
    print(tree)   # ASCII আকারে tree দেখাবে

    # Matplotlib দিয়ে আঁকা
    def draw_tree(tree, x=0, y=0, ax=None):
        if ax is None:
            fig, ax = plt.subplots(figsize=(6, 4))
            ax.axis('off')

        ax.text(x, y, tree.label(), ha='center', va='center',
                bbox=dict(facecolor='lightblue', edgecolor='black'))

        if isinstance(tree, Tree):
            step = 1.0 / (len(tree) + 1)
            for i, child in enumerate(tree):
                child_x = x + (i + 1 - len(tree)/2) * step * 6
                child_y = y - 1.5
                ax.plot([x, child_x], [y-0.2, child_y+0.2], color='black')
                if isinstance(child, Tree):
                    draw_tree(child, child_x, child_y, ax=ax)
                else:
                    ax.text(child_x, child_y, child, ha='center', va='center',
                            bbox=dict(facecolor='lightgreen', edgecolor='black'))
        return ax

    ax = draw_tree(tree)
    plt.show()
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk.tree import Tree

# Example CFG বানানো
cfg = nltk.CFG.fromstring("""
S -> A 'a'
A -> 'b' | 'b' A | 'b' A 'c'
""")

# Input text
text = "b a"
words = text.split()

# Parser তৈরি
parser = nltk.ChartParser(cfg)

# Parse করে tree পাওয়া
for tree in parser.parse(words):
    print(tree)          # ASCII আকারে tree দেখাবে
    tree.pretty_print()  # console view

    # Matplotlib দিয়ে আঁকার function
    def draw_tree(tree, x=0, y=0, ax=None):
        if ax is None:
            fig, ax = plt.subplots(figsize=(8, 6))
            ax.axis('off')

        ax.text(x, y, tree.label(), ha='center', va='center',
                bbox=dict(facecolor='lightblue', edgecolor='black'))

        if isinstance(tree, Tree):
            step = 1.0 / (len(tree) + 1)
            for i, child in enumerate(tree):
                child_x = x + (i + 1 - len(tree)/2) * step * 10
                child_y = y - 1.5
                ax.plot([x, child_x], [y-0.2, child_y+0.2], color='black')
                if isinstance(child, Tree):
                    draw_tree(child, child_x, child_y, ax=ax)
                else:
                    ax.text(child_x, child_y, child, ha='center', va='center',
                            bbox=dict(facecolor='lightgreen', edgecolor='black'))
        return ax

    ax = draw_tree(tree)
    plt.show()
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
text='NLTK is a powerful library for natural language processing.'
words=word_tokenize(text)
pos_tags=pos_tag(words)
print("Original Text:")
print(text)
print('word tokenization: ',words)

print('pos tagging result: ')
for word,pos_tag in pos_tags:
  print(f'{word} : {pos_tag}')
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
text='Spacy is a popular natural language processing library.'
doc=nlp(text)
print('Tokenization:')
for token in doc:
    print(token.text)
print('Original Text:',text)
print('Pos tagging result')
for token in doc:
    print(f'{token.text} : {token.pos_}')
    print('Original Text: ',text)
```
