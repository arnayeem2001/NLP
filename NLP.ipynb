# Vocabulary and Matching

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import Matcher
matcher=Matcher(nlp.vocab)
```

```python
pattern_1=[{'LOWER':'hello'},{'LOWER':'world'}]
pattern_2=[{'LOWER':'hello'},{'IS_PUNCT':True},{'LOWER':'world'}]
```

```python
matcher.add("HelloWorld", [pattern_1, pattern_2])
```

```python
doc=nlp("HelloWorld hello world hello-world")
```

```python
doc
```

```python
find_matches=matcher(doc)
print(find_matches)
```

```python
for match_id, start, end in find_matches:
  string_id=nlp.vocab.strings[match_id]
  span=doc[start:end]
  print(match_id, string_id, start, end, span.text)
```

```python
matcher.remove('HelloWorld')
```

```python
pattern_3=[{'LOWER':'hello'},{'LOWER':'world'}]
pattern_4=[{'LOWER':'hello'},{'IS_PUNCT':True, 'OP':'*'},{'LOWER':'world'}]
```

```python
matcher.add("HelloWorld", [pattern_3, pattern_4])
```

```python
doc_2=nlp("You can print Hello World or hello world or Hello-World")
```

```python
find_matches=matcher(doc_2)
print(find_matches)
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import PhraseMatcher
matcher=PhraseMatcher(nlp.vocab)
```

```python
phrase_list=['Hasina','Tarek','Ershad']
```

```python
phrase_patterns=[nlp(text) for text in phrase_list]
```

```python
phrase_patterns
```

```python
matcher.add('TerminologyList',None,*phrase_patterns)
```

```python
doc_3=nlp('Hasina is the greatest terorist in bd history'
        'after 2024 in bd, bnp pratices ilegal procedure')
```

```python
find_matches=matcher(doc_3)
print(find_matches)
```

```python
for match_id, start, end in find_matches:
  string_id=nlp.vocab.strings[match_id]
  span=doc_3[start:end]
  print(match_id, string_id,start,end,span.text)
```
